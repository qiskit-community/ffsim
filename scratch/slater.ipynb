```python
import ffsim
from qiskit.circuit import QuantumCircuit, QuantumRegister
```

```python
norb = 8
nelec = 2

orbital_rotation = ffsim.random.random_unitary(norb)

qubits = QuantumRegister(norb)
circuit = QuantumCircuit(qubits)
circuit.append(
    ffsim.qiskit.PrepareSlaterDeterminantSpinlessJW(
        norb, range(nelec), orbital_rotation=orbital_rotation
    ),
    qubits,
)
# (8 - 3) * 3, nf => nelec, n -> norb
# (norb - nelec) + max_gate_layer

circuit.decompose().draw("mpl", scale=0.7)
```
